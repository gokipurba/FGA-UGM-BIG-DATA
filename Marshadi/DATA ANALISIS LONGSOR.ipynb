### inti dari machine learning proses model dari data
### Training Range 70 - 90 %


```python
from sklearn.model_selection import train_test_split
```

```python
y = df["Quality_type"]

X = df.drop(["DateTime", "Quality_type"], axis= 1)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.15,random_state = 15)


print('X_train = ', X_train.shape)
print('X_test = ', X_test.shape)
print('y_train = ', y_train.shape)
print('y_test = ', y_test.shape)
```

# Model Creation, Training, Evaluation

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
# Evaluation
# Compare prediction

df["Quality_prediction"] = rfc.predict(X)
df.sample(15,random_state = 22)
```

```python
rfc.feature_importances_
```

```python
rfc_imp = pd.DataFrame(rfc.feature_importances_, columns = ["Importance"])
rfc_imp["Importance"] = rfc_imp["Importance"]*100

rfc_imp = rfc_imp.set_index([X.columns])

display(rfc_imp.sort_values(by="Importance",ascending = False))
                    
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNC = KNeighborsClassifier(n_neighbors = 3)
```

```python
KNC.fit(X_train, y_train)
```

```python
print("Train = ",KNC.score(X_train, y_train))
print("Test = ",KNC.score(X_test, y_test))
```

```python
# Evaluation
# Compare prediction

df["Quality_prediction"] = KNC.predict(X)
df.sample(15,random_state = 22)
```
